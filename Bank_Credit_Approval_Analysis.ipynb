```python

# Find outliers for numerical columns
numerical_columns = ['CNT_CHILDREN', 'AMT_INCOME_TOTAL', 'AGE', 'EXP_IN_YEARS', 'CNT_FAM_MEMBERS']
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5* IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col} outliers: {len(outliers)}")
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
principle_amount = 1000

def calculate_bank_cost(row):
    predicted = row['Predicted']
    actual = row['Actual']
    status = row['STATUS']
    bank_cost = 0
    
    if predicted == 0 and actual == 1 and status == 0:
        bank_cost = principle_amount * 0.04 * 1 * -1
    elif predicted == 0 and actual == 1 and status == 1:
        bank_cost = principle_amount * 0.04 * 2 * -1
    elif predicted == 0 and actual == 1 and status == 2:
        bank_cost = principle_amount * 0.04 * 3 * -1
    elif predicted == 0 and actual == 1 and status == 3:
        bank_cost = principle_amount * 0.04 * 4 * -1
    elif predicted == 0 and actual == 1 and status == 4:
        bank_cost = principle_amount * 0.04 * 5 * -1
    elif predicted == 0 and actual == 1 and status == 5:
        bank_cost = principle_amount * 0.04 * 6 * -1
    elif predicted == 1 and actual == 1:
        bank_cost = 0
    elif predicted == 1 and actual == 0 and status == 1:
        bank_cost = principle_amount * 0.04 * 2
    elif predicted == 1 and actual == 0 and status == 2:
        bank_cost = principle_amount * 0.04 * 3
    elif predicted == 1 and actual == 0 and status == 3:
        bank_cost = principle_amount * 0.04 * 4
    elif predicted == 1 and actual == 0 and status == 4:
        bank_cost = principle_amount * 0.04 * 5
    elif predicted == 1 and actual == 0 and status == 5:
        bank_cost = principle_amount * 0.04 * 6
    elif predicted == 0 and actual == 0:
        bank_cost = 0    
    else:
        bank_cost = principle_amount * 0.04
        
    return bank_cost

analysis['Bank Cost'] = analysis.apply(calculate_bank_cost, axis=1)
```
